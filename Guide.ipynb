# Финальный проект первого семестра DLSchool. 

## Style transfer telegram bot.

В данном проекте мы хотим сделать телеграм бота, который по двум картинкам будет переносить стиль с одной на другую.

В данной проекте у нас есть две папки:
* model - папка для ноутбуков с экспериментами. Именно в этой части проекта мы будем строить свою модель, обучать ее и тестировать.
* telegram_bot - папка с .py файлами, которые относятся к продакшн части нашего проекта. Здесь будет лежать наш файл с телеграм ботом, а так же перенесенная из тетрадки модель, вся работа которой скрыта в классе.

У нашего проекта есть небольшая особенность - экспериментальной части в начале не будет, из-за того, что модель уже была написана на занятии и есть в тетрадке. Вам нужно будет только перенести ее. Но мы здесь собрались ради машинного обучения! Поэтому, если вы быстро закончите бота с этой базовой версией style transfer, то стоит попробовать реализовать более продвинутый алгоритм, который работает нанмого быстрее. Больше информации и ссылки на статьи будут позже, когда я сам до этого дойду или вы попросите меня их найти)

В бейзлайне используется библиотека для ботов python telegram bot. Документация к ней легко гуглится. 

Если при запуске бейзлайна вам не хватает какой-то библиотеки - просто забивайте в гугл название отсутствующего модуля и pip'ом устанавливайте нужную библиотеку.

Думаю, что наша работа будет максимально продуктивной, если вы будете двигаться по следющему курсу:

0. Написать BotFather и получить свой токен. В папке telegram_bot создать файл telegram_token.py с единственной строчкой: token = 'ВАШ ТОКЕН ЗДЕСЬ'. **Будьте осторожны и не загрузите свой токен на гитхаб!!!** Я уже включил файл telegram_token.py в .gitignore, поэтому токен не попадет на гитхаб, если вы будете держать его только в этом файле.
1. Созддайте свой **гитхаб репозиторий** и сделайте начальный коммит, состоящий из файлов бейзлайна, и скиньте ссылку в слак, чтобы я имел доступ к репозиторию. Эту тетрадку включать не стоит, для удобства я внесу её в .gitignore. Если вы не можете в гит, то напишите мне, я помогу разобраться.
2. Перенесите **базовый алгоритм** style transfer из тетрадки с занятия в питоновский класс, который уже определен в бейзлайне.
3. Доработайте самого **бота**: **а**) добавьте приветствие и объяснение принципа работы, сделайте кнопки, чтобы человек мог переходить в режим стайл трансфера и выходить из него, видеть, отправил ли он уже первую картинку (Такая необходимость связана с тем, что в телеграме каждая картинка приходит отдельно и человек может отправить первую, а потом захотеть сбросить это действие) и т. д. Будет хорошо, если все строчки, которые отправляет бот вы вынесете в отдельный файл. Например, config.py. **б**) Придумайте, как сделать так, чтобы через какое-то время информация о первой отправленной картинке исчезала. То есть чтобы была некоторая сессия, которая исчезает через 15 минут  неактивности.
4. Теперь, когда бот уже полностью рабочий и никаких изменений в него не планируется, можно перейти к реализации более **сложного алгоритма**, для этого вам и понадобится папка model, где вы будете сами писать модель по статье и заметкам на medium). Более подробные объяснения я дам, когда вы будете близки к этому этапу.
5. Когда сложный алгоритм заработает, его тоже нужно будет **перенести в свой класс** (скорее всего стоит заменить код простого алгоритма) и снова **заставить бота правильно работать**.
6. Теперь все готово и остались только **косметические изменения**: добавление docstring, исправение кодстайла, рефакторинг и т.д. 
7. ...
8. PROFIT!!!
